## Determination of the bandwidth of the ultrasonic transducer

### Measurement of Resonant Frequency

```python
red = {X: [41.12, 41.25, 41.07, 41.05, 41.02, 41.19, 41.23],
       Y: [2820, 2800, 2800, 2840, 2880, 2840, 2820]}
re = pd.DataFrame(red)
```

```python
display(Latex("Measured resonant frequency, $\overline f = ({} \pm {:10.2f} ) \ kHz$".format(41.132857, re.sem(axis=0)[0])))
display(Latex("Measured amplitude at resonant frequency, $\overline A = ({} \pm {:10.2f} ) \ mV$".format(2828.57, re.sem(axis=0)[1])))
```

```python
f = codecs.open('data11.txt').read().replace('\t', ',')
i = open('new.txt', "w")
i.write(f)
i.close()
p1 = pd.read_csv('new.txt', names=[X, Y])
p1
```

```python
p1.describe()
```

```python
X_np = np.array(p1[X])
Y_np = np.array(p1[Y])
model = LorentzianModel()
params = model.guess(X_np, x=Y_np)

result = model.fit(Y_np, params, x=X_np)
result.plot_fit()
plt.xlabel("Frequency [$kHz$]")
plt.ylabel('Amplitude [$mV$]')
plt.title("Frequency vs Amplitude for Transducer Pair (1) and (3)")
plt.legend('')
plt.grid()
plt.show()
```

## Determination of acoustic velocity with the sonic depth finder

```python
display(Latex("The first train wave is due to the transmitter, and it was detected early by the receiver being very near to it. \
The second train wave, which is detected later, is due the reflection wave from the reflecting plate."))
```

```python
X = "Distance [$m$]"
Y = "Time [$s$]"
f = codecs.open('re.txt').read().replace('\t', ',')
i = open('new.txt', "w")
i.write(f)
i.close()
p3 = pd.read_csv('new.txt', names=[X, Y])
p3[X] = (2 * (p3[X] - 30))/(1000)
p3[Y] = (p3[Y]) / 1000
p3
```

```python
print(p3.to_latex(index=True, escape=False))
```

```python
B1 = np.array(p3[X]).reshape(-1, 1)
A = np.array(p3[Y]).reshape(-1, 1)
reg = LinearRegression().fit(A, B1)
Y_pred1 = reg.predict(A)
reg_value = reg.score(A, B1)
intercept = reg.intercept_
coef = reg.coef_
error = np.sqrt((1/(p3[X].describe().loc['count'] - 2)) * (1 - ((reg_value) ** 2))/((reg_value) ** 2)) * coef

plt.scatter(A, B1)
plt.plot(A, Y_pred1, color='blue')
plt.xlabel(Y)
plt.ylabel(X)
plt.xlim(0.0009, 0.006)

plt.grid()
plt.show()
display(Latex("$R^2$ = {:10.4f}".format(reg_value)))
display(Latex("$f(x) = {:10.2f}x  {:10.2f}$".format(coef[0, 0], intercept[0])))
display(Latex("$\Delta m = {:10.2f}$".format(error[0][0])))
```

```python
display(Latex("The velocity of sound is measured to be $(346.31 \pm 1.95) \ ms^{-1}.$"))
```

#### The reflecting plate is removed, and the transmitter and receiver are faced each other.

```python
X = "Distance [$m$]"
Y = "Time [$s$]"
f = codecs.open('wore.txt').read().replace('\t', ',')
i = open('new.txt', "w")
i.write(f)
i.close()
p3 = pd.read_csv('new.txt', names=[X, Y])
p3[X] = ((p3[X] - 60))/(1000)
p3[Y] = (p3[Y]) / 1000
p3
```

```python
print(p3.to_latex(index=True, escape=False))
```

```python
B2 = np.array(p3[X]).reshape(-1, 1)
A = np.array(p3[Y]).reshape(-1, 1)
reg = LinearRegression().fit(A, B2)
Y_pred2 = reg.predict(A)
reg_value = reg.score(A, B2)
intercept = reg.intercept_
coef = reg.coef_
error = np.sqrt((1/(p3[X].describe().loc['count'] - 2)) * (1 - ((reg_value) ** 2))/((reg_value) ** 2)) * coef

plt.scatter(A, B2)
plt.plot(A, Y_pred2, color='blue')
plt.xlabel(Y)
plt.ylabel(X)
plt.xlim(0.0003, 0.003)
plt.grid()
plt.show()
display(Latex("$R^2$ = {:10.4f}".format(reg_value)))
display(Latex("$f(x) = {:10.2f}x  {:10.2f}$".format(coef[0, 0], intercept[0])))
display(Latex("$\Delta m = {:10.2f}$".format(error[0][0])))
```

```python
display(Latex("The velocity of sound is measured to be $(332.35 \pm 2.43) \ ms^{-1}.$"))
```
